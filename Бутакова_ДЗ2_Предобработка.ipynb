```python
from nltk import word_tokenize
```

```python
# 1 Выполните токенизацию нормализованного текста на слова. Сколько токенов в тексте? А сколько уникальных?

tokens = word_tokenize(text_clean)
print(f'Количество токенов в тексте: {len(tokens)}.\nУникальных токенов: {len(set(tokens))}.')
```

```python
#2. Выполните токенизацию изначального текста на строки при помощи split('\n'). Отобразите сколько их.

strings = text.split('\n')
strings_cut = list(filter(None, strings)) # есть пустые элементы там, где были пустые строки - удалим их
print(f'Количество строк после разбивания по "\\n": {len(strings)}. \nПосле удаления пустых строк: {len(strings_cut)}.')
```

```python
# 3. Напишите цикл, который пройдет по всем токенам текста и сохранит в новой переменной те, которые начинаются с заглавной буквы.

# у нас были нормализованные токены-слова, а теперь нужно, чтобы сохранялся регистр - подготовим токены для этого задания
tokens_raw = re.sub(r'[\.,!?:;—]', '', text).split() # очистили от пунктуации + разбили по пробелу

tokens_upper = []
for t in tokens_raw:
    if t[0].isupper():
        tokens_upper.append(t)

print(tokens_upper)
```

```python
# 4. *Модернизируйте цикл так, чтобы он сохранял в переменную слова с заглавной буквы за исключением тех, которые являются первыми в строке.

upper_notfirst = []
for t in tokens_raw:
    if t[0].isupper():
        if re.search(fr'(?<!\n){t}', text): # проверяем слова с заглавной с помощью negative lookbehind 
            upper_notfirst.append(t) # если перед словом нет переноса строки, нам подходит

print(upper_notfirst)
        
```

Задание 4: Удаление стоп слов


1. Импортируйте список стоп-слов для русского языка из NLTK.
2. Напишите цикл, который пройдет по всем токенам и вернет новый список токенов без стоп-слов.
3. Сколько слов осталось в тексте? А уникальных?
4. *При помощи регулярного выражения найдите все слова с сочетанием букв 'бол' (с опциональной о) внутри слова. Добавьте их к стоп-словам и удалите из текста. Сколько осталось слов в тексте?




```python
# 1. Импортируйте список стоп-слов для русского языка из NLTK.

from nltk.corpus import stopwords 
stopwords_ru = stopwords.words('russian')
```

```python
tokens
```

```python
# 2. Напишите цикл, который пройдет по всем токенам и вернет новый список токенов без стоп-слов.

tokens_nostop = []
for t in tokens: # мы уже ранее все токенизировали, воспользуемся этим
    if t not in stopwords_ru:
        tokens_nostop.append(t)
print(tokens_nostop)
```

```python
# 3. Сколько слов осталось в тексте? А уникальных?
print(f'Количество токенов после удаления стоп-слов: {len(tokens_nostop)} (до удаления их было {len(tokens)}).\nУникальных токенов: {len(set(tokens_nostop))}.')
```

```python
# 4. *При помощи регулярного выражения найдите все слова с сочетанием букв 'бол' (с опциональной о) внутри слова. 
# Добавьте их к стоп-словам и удалите из текста. Сколько осталось слов в тексте?

to_exclude = re.findall(r'\b\w*?бо?л\w+?\b', ' '.join(tokens_nostop))
stopwords_new = stopwords_ru + to_exclude

tokens_nostop_extended = []
for t in tokens:
    if t not in stopwords_new:
        tokens_nostop_extended.append(t)
print(f'Количество токенов после удаления добавочных стопслов: {len(tokens_nostop_extended)}.\nУдаленные слова: {set(tokens_nostop) - set(tokens_nostop_extended)}')

```
